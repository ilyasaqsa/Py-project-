```python
class Hospital:

    def __init__(self):
        self.patient_name = ''
        self.patient_age = 0
        self.appointment = ''
        self.menu()

    def menu(self):
        print("---------------------------------------") 
        user_input = input("""
        
 Welcome to Hospital Appointment System 
        
 1. Press 1 to register patient
 2. Press 2 to book appointment
 3. Press 3 to cancel appointment
 4. Press 4 to view appointment
 5. Press anything else to exit
        """)

        if user_input == '1':
            self.register_patient()
        elif user_input == '2':
            self.book_appointment()
        elif user_input == '3':
            self.cancel_appointment()
        elif user_input == '4':
            self.view_appointment()
        else:
            print("\nThank you! Stay healthy.\n")
            exit()

    def register_patient(self):
        print("---------------------------------------")
        self.patient_name = input("Enter patient name: ")
        self.patient_age = input("Enter patient age: ")
        print("\nPatient registered successfully\n")
        self.menu()

    def book_appointment(self):
        print("---------------------------------------")
        if self.patient_name == '':
            print("Please register patient first\n")
        else:
            self.appointment = input("Enter doctor name / department: ")
            print("\nAppointment booked successfully\n")
        self.menu()

    def cancel_appointment(self):
        print("---------------------------------------")
        if self.appointment == '':
            print("No appointment to cancel\n")
        else:
            self.appointment = ''
            print("\nAppointment cancelled successfully\n")
        self.menu()

    def view_appointment(self):
        print("---------------------------------------")
        if self.appointment == '':
            print("No appointment booked\n")
        else:
            print("Patient Name:", self.patient_name)
            print("Patient Age:", self.patient_age)
            print("Appointment with:", self.appointment)
            print()
        self.menu()
        
h = Hospital()
```
