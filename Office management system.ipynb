```python
# office_management_system.py

# -------------------- CLASSES -------------------- #
class Employee:
    def __init__(self, emp_id, name, age, role, dept):
        self.emp_id = emp_id
        self.name = name
        self.age = age
        self.role = role
        self.dept = dept

    def display(self):
        print(f"ID: {self.emp_id} | Name: {self.name} | Age: {self.age} | Role: {self.role} | Dept: {self.dept}")


class Department:
    def __init__(self, dept_id, name):
        self.dept_id = dept_id
        self.name = name

    def display(self):
        print(f"ID: {self.dept_id} | Name: {self.name}")


class Task:
    def __init__(self, task_id, title, emp_id, status="Pending"):
        self.task_id = task_id
        self.title = title
        self.emp_id = emp_id
        self.status = status

    def complete(self):
        self.status = "Completed"

    def display(self):
        print(f"Task ID: {self.task_id} | Title: {self.title} | Emp: {self.emp_id} | Status: {self.status}")


class Attendance:
    def __init__(self, emp_id, date, status):
        self.emp_id = emp_id
        self.date = date
        self.status = status

    def display(self):
        print(f"Emp ID: {self.emp_id} | Date: {self.date} | Status: {self.status}")


# -------------------- DATA -------------------- #
employees = []
departments = []
tasks = []
attendances = []


# -------------------- FUNCTIONS -------------------- #
def add_employee():
    employees.append(Employee(input("ID: "), input("Name: "), input("Age: "), input("Role: "), input("Dept: ")))
    print("Employee added!\n")


def view_employees():
    if employees:
        for e in employees:
            e.display()
    else:
        print("No employees!\n")


def search_employee():
    emp_id = input("Employee ID: ")
    for e in employees:
        if e.emp_id == emp_id:
            e.display()
            return
    print("Employee not found!\n")


def add_department():
    departments.append(Department(input("ID: "), input("Name: ")))
    print("Department added!\n")


def view_departments():
    if departments:
        for d in departments:
            d.display()
    else:
        print("No departments!\n")


def assign_task():
    tasks.append(Task(input("Task ID: "), input("Title: "), input("Emp ID: ")))
    print("Task assigned!\n")


def view_tasks():
    if tasks:
        for t in tasks:
            t.display()
    else:
        print("No tasks!\n")


def complete_task():
    tid = input("Task ID: ")
    for t in tasks:
        if t.task_id == tid:
            t.complete()
            print("Task completed!\n")
            return
    print("Task not found!\n")


def mark_attendance():
    attendances.append(Attendance(input("Emp ID: "), input("Date: "), input("Status: ")))
    print("Attendance marked!\n")


def view_attendance():
    if attendances:
        for a in attendances:
            a.display()
    else:
        print("No attendance records!\n")


# -------------------- MENUS -------------------- #
def menu(submenu):
    while True:
        choice = input(submenu)
        if choice == "1":
            if "Employee" in submenu:
                add_employee()
            elif "Department" in submenu:
                add_department()
            elif "Task" in submenu:
                assign_task()
            elif "Attendance" in submenu:
                mark_attendance()

        elif choice == "2":
            if "Employee" in submenu:
                view_employees()
            elif "Department" in submenu:
                view_departments()
            elif "Task" in submenu:
                view_tasks()
            elif "Attendance" in submenu:
                view_attendance()

        elif choice == "3":
            if "Employee" in submenu:
                search_employee()
            elif "Task" in submenu:
                complete_task()
            else:  
                break

        elif choice == "4":
            if "Employee" in submenu or "Department" in submenu or "Task" in submenu:
                break
            else:
                print("Invalid choice!\n")
        else:
            print("Invalid choice!\n")

        if "Attendance" in submenu and choice == "3":
            break


def report_menu():
    while True:
        print("1. Employee\n2. Department\n3. Task\n4. Attendance\n5. Back")
        c = input("Choice: ")
        if c == "1":
            view_employees()
        elif c == "2":
            view_departments()
        elif c == "3":
            view_tasks()
        elif c == "4":
            view_attendance()
        elif c == "5":
            break
        else:
            print("Invalid choice!\n")


# -------------------- MAIN MENU -------------------- #
def main_menu():
    while True:
        print("\n=== OFFICE MANAGEMENT SYSTEM ===")
        print("1. Employee Management")
        print("2. Department Management")
        print("3. Task Management")
        print("4. Attendance System")
        print("5. Reports")
        print("6. Exit")
        choice = input("Choice: ")
        if choice == "1":
            menu("Employee Management:\n1.Add\n2.View\n3.Search\n4.Back\nChoice: ")
        elif choice == "2":
            menu("Department Management:\n1.Add\n2.View\n3.Back\nChoice: ")
        elif choice == "3":
            menu("Task Management:\n1.Assign\n2.View\n3.Complete\n4.Back\nChoice: ")
        elif choice == "4":
            menu("Attendance Management:\n1.Mark\n2.View\n3.Back\nChoice: ")
        elif choice == "5":
            report_menu()
        elif choice == "6":
            print("Exiting system. Goodbye!")
            break
        else:
            print("Invalid choice!\n")


# -------------------- RUN -------------------- #
if __name__ == "__main__":
    main_menu()
```
